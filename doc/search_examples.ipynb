```python
filename = './all-the-news/articles1.csv'
df = all_news_filereader(filename)
```

```python
# limit to the first 1000 news
# df = df[0:1000]
```

build models for **title**, **publication**, **author**, **content** columns

```python
models = build_models(df, usecols=['title', 'publication', 'author', 'content'])
```

```python
def simple_search(query):
    parsed_query = parse_input(query)
    return ranker(df, models, parsed_query)
```

perform and display the queries requested in the challenge

```python
r1 = simple_search('Trump')
r1
```

```python
r1.to_csv('results_trump.csv')
```
